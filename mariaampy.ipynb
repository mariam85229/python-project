```python
class bank:
    def __init__(self):
        self.bank_acc={}
        self.bank_balance={}  
        
    def register(self,username,password):
        if username in self.bank_acc.keys():
            print("Username already exists. Please choose a different username.")
        else:
            self.bank_acc[username]=password
            self.bank_balance[username]=0
            print("successfully registered")

    def login(self,username,password):
        if username in self.bank_acc.keys() and password in self.bank_acc.values():
            print("login successfully")
            return username     
        else:
            print("incorrect username or password")
    
    def acc_operation(self,username):
        if username in self.bank_balance.keys() :
            print(f"your current balance is : {self.bank_balance[username]}")

    def deposit(self,amount,username):
        amount=int(amount)
        if amount<0:
            return("cannot add")
        
        if username in self.bank_balance.keys() :
           self.bank_balance[username]+=amount
           print(f"current balance is {self.bank_balance[username]}")
    def withdraw(self,username,amount):
        amount=int(amount)
        if amount > self.bank_balance[username] :
           return("cannot withdraw")
        amount=int(amount)
        self.bank_balance[username]-=amount
        print(f"after withdraw your balance is {self.bank_balance[username]}")

bank1=bank()

while True:
    print("welcome to our bank please register or login")
    action=input("enter your action")
    if action=="register":
        username=input("user for reg ")
        password=input("pass for reg")
        bank1.register(username,password)
    elif action=="exit":
        break
    elif action== "login":
        username=input("user for login")
        password=input("pass for login")
        user_name=bank1.login(username,password)
        while user_name in bank1.bank_acc.keys() :
             print("choose deposit or withdraw or balance checking")
             check=input("check")
             if check=="deposit":
                 
                 amount=int(input("amount for dep"))
                 bank1.deposit(amount,user_name)
             elif check=="balance checking":
                 
                 bank1.acc_operation(user_name)
             elif check=="withdraw":
                
                amount=int(input("amount for withdraw"))
                bank1.withdraw(user_name, amount)

             elif check =="exit":
                 break   


```
